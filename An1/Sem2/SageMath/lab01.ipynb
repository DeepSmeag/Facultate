```python
t = var('t')
f(t,s) = 1- (s*cos(4*t)*cos(t))/(1-s^2*cos(4*t)^2*sin(t)^2)^(1/2)
x(t,s) = f(t-pi/2,s)
y(t,s) = f(t,s)
g= parametric_plot((x(t,1/10),y(t,1/10)),(t,0,2*pi))
for s in [2..10]:
    g1 = parametric_plot((x(t,s/10),y(t,s/10)),(t,0,2*pi))
    g = g+g1
show(g)
```
