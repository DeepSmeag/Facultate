```python
reset()
t = var('t')
omegazero = var('omegazero')
lbd = var('lbd')
x = function('x')(t)
assume(omegazero>0)
assume(lbd^2 > 4 * omegazero^2)
eqd = diff(x,t,2) + lbd * diff(x,t) + omegazero^2 * x == 0
k1 = var('k1')
k2 = var('k2')
sol(t, k1, k2) = desolve(eqd, [x,t])

solcond(t, lbd, omegazero) = desolve(eqd, [x,t], [0,1,5])
solt(t) = solcond(t, 25, 10)
T = var('T')
T = 2 * pi / 100

plot(solt(t), t, 0, 10*T)
forget()
assume(lbd>0)
assume(omegazero>0)
eqd2 = diff(x,t,2) + 2*omegazero*diff(x,t) + omegazero^2 * x==0
soltcrit(t, omegazero, k1, k2) = desolve(eqd2, [x,t])
# show(soltcrit)
solcond2(t, omegazero) = desolve(eqd2, [x,t], [0,1,5])
solt2(t) = solcond2(t,10)
plot(solt2(t), t, 0, 1)

forget()
assume(lbd^2 < 4 * omegazero^2)
solslab(t, lbd, omegazero, k1, k2) = desolve(eqd, [x,t])
# show(solslab)
solcond3(t, lbd, omegazero) = desolve(eqd, [x,t], [0,1,5])
# show(solcond3)
solt3(t) = solcond3(t, 5, 10)
# show(solt3)
plot(solt3(t),t,0,1, detect_poles='True')
```
